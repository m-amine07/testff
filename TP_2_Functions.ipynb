```python
def is_adult(age):
  if age >= 18:
    return True
  else:
    return False
```

```python
def display_age_status(age):
  if is_adult(age):
    print("majeur")
  else:
    print("mineur")
display_age_status(15)
```

```python
def format_name(first, last):
  return first.capitalize() + " " + last.capitalize()
format_name("amine", "kbaier")
```

```python
def greet_user(first,last):
  print(f"bonjour, {format_name(first,last)}")
greet_user("amine", "kbaier")
```

```python
def avg(a,b,c):
  return (a+b+c)/3
avg(1,2,3)

```

```python
def finale_grade(a,b,c):
  c=(a*avg(a,b,c))+(b*avg(a,b,c))+(c*avg(a,b,c))
  if c>=16:
    return "EXCELLENT"
  elif c>=12:
    return "BIEN"
  else:
    return "insuffisant"
finale_grade(2,5,3)

```

```python
def clean_text(text):
  c=".,!?"
  for i in c :
    text=text.replace(i,"")
  return text.lower().strip()
clean_text(" AMine! kbaier. ")

```

```python
def count_words(text):
  worlds=clean_text(text).split()
  return len(worlds)


count_words("amine kbaier")
```

```python
def apply_discount(price, precent):
  return price - (price * precent / 100)
apply_discount(120, 15)
```

```python
def best_price(prices):
  return list(map(lambda a: apply_discount(a, 10),prices))
best_price([12,13,15])
```

```python
def has_number(text):
  for i in text:
    if i.isdigit():
      return True
  return False
has_number("amine00")
```

```python
def has_special(text):
  for i in text:
    if i in '.!?*@':
      return True
  return False
has_special("amine")
```

```python
def password_is_valid(password):
  if has_number(password) and has_special(password):
    return "Mot de passe valide"
  else:
    return "mot e passe invalide"

password_is_valid("ami!ne00")
```

```python
def extract_digits(num):
  return list(map(int,str(num)))
extract_digits(12345)
```

```python
def sum_even_digtis(num):
  return sum(extract_digits(num))
sum_even_digtis(12345)
```
